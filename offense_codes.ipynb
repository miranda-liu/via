```python
import pandas as pd

df = pd.read_excel('/Users/emilyzhou/Desktop/rmsoffensecodes.xlsx')
codelist = df['CODE'].tolist()
offense = df['NAME'].tolist()
code_list_new = []
offenselist_new = []

for elt in offense:
    if pd.isna(elt) == False:
        offenselist_new.append(elt)

for elt in codelist:
    if pd.isna(elt) == False:
        code_list_new.append(elt)
        
key_value_pairs = zip(code_list_new, offenselist_new)

# convert the list of key-value pairs to a dictionary
my_dict = dict(key_value_pairs)

print(my_dict)
```
